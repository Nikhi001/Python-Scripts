```python
import os
import pytesseract
from pdf2image import convert_from_path
from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException

# Your folder path
folder_path = '/content/drive/MyDrive/ImmverseAI Project/Data/PDF Book'

# Get all PDF files
pdf_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.pdf')]

# Make sure these are initialized
ocr_data = {}
detected_languages = {}

for filename in pdf_files:
    pdf_path = os.path.join(folder_path, filename)
    print(f" Processing: {filename}")

    try:
        # Convert first 5 pages to images
        images = convert_from_path(pdf_path, first_page=1, last_page=5)

        page_texts = {}
        for i, img in enumerate(images):
            text = pytesseract.image_to_string(img, lang='hin+san+kas+pli+bra+awa+nep+mar+kok+eng')
            page_texts[f"Page_{i + 1}"] = text.strip()

        # Save OCR result
        ocr_data[filename] = page_texts

        # Language Detection
        full_text = " ".join(page_texts.values())
        try:
            lang = detect(full_text)
        except LangDetectException:
            lang = "Unknown"
        detected_languages[filename] = lang

    except Exception as e:
        print(f"Error processing {filename}: {e}")
        detected_languages[filename] = "Error"

# Print sample output
print("\n Sample OCR Output:")
for file, pages in list(ocr_data.items())[:2]:
    print(f"\n {file}")
    for pg, text in pages.items():
        print(f"{pg}:", text[:150], "...")  # print first 150 characters

#  Print detected languages
print("\n Detected Languages:")
for file, lang in detected_languages.items():
    print(f"{file}: {lang}")
```

```python
import re
import pandas as pd

# Keywords
author_keywords = ['author','Author', 'लेखक', 'कर्ता']
editor_keywords = ['editor','Editor', 'सम्पादक', 'सम्पादन']
publisher_keywords = ['publisher','Publisher', 'published by','Published by', 'प्रकाशक', 'प्रकाशन']
year_keywords = ['year', 'वर्ष', 'सन्', 'प्रकाशन वर्ष']
year_patterns = [r'\b(?:19|20)\d{2}\b']

# Lists to collect extracted data
file_names = []
authors = []
editors = []
publishers = []
years = []
languages = []

# Loop through `ocr_data` instead of text
for filename, pages in ocr_data.items():
    full_text = " ".join(pages.values()).lower() # join all text from pages in a string convert into lower case
    content_clean = re.sub(r'[–—\-_:|•]+', ':', full_text) # help to find with key values

    # Defaults use if any value not found from string will fill with "Unknown"
    author = "Unknown"
    editor = "Unknown"
    publisher = "Unknown"
    year = "Unknown"

    # Extract author with help of keywords
    for kw in author_keywords:
        match = re.search(rf'{kw}\s*[:]\s*(.*)', content_clean)
        if match:
            author = match.group(1).split('\n')[0].strip(" :–—,")
            break

    # Extract editor with help of keywords
    for kw in editor_keywords:
        match = re.search(rf'{kw}\s*[:]\s*(.*)', content_clean)
        if match:
            editor = match.group(1).split('\n')[0].strip(" :–—,")
            break

    # Extract publisher with help of keywords
    for kw in publisher_keywords:
        match = re.search(rf'{kw}\s*[:]\s*(.*)', content_clean)
        if match:
            publisher = match.group(1).split('\n')[0].strip(" :–—,")
            break

    # Extract year same with help of year pattern and
    for kw in year_keywords:
        match = re.search(rf'{kw}\s*[:]\s*(\d{{4}})', content_clean)
        if match:
            year = match.group(1)
            break
    if year == "Unknown": # fallback technique used
        for pattern in year_patterns:
            match = re.search(pattern, content_clean)
            if match:
                year = match.group(0)
                break
    if year == "Unknown": # fallback technique used on filename
        match = re.search(r'(19|20)\d{2}', filename)
        if match:
            year = match.group(0)

    # Append all
    file_names.append(filename)
    authors.append(author)
    editors.append(editor)
    publishers.append(publisher)
    years.append(year)
    languages.append(detected_languages.get(filename, "Unknown"))
```

```python
# Create final dataframe
df_final = pd.DataFrame({
    "File Name": file_names,
    "Author": authors,
    "Editor": editors,
    "Publisher": publishers,
    "Year of Publication": years,
    "Language": languages
})


df_final
```

```python
import pandas as pd

# Merge first
df_merged = pd.merge(df, df_final, on='File Name', how='left', suffixes=('_x', '_y'))

# Define function to combine values with logic
def combine_values(x, y):
    if pd.isna(x) or str(x).strip().lower() == "unknown" or str(x).strip() == "":
        return y
    if str(x).strip() == str(y).strip():
        return x
    return x  # x is preferred if present and different

# Combine overlapping fields
df_merged['Author'] = df_merged.apply(lambda row: combine_values(row.get('Author_x'), row.get('Author_y')), axis=1)
df_merged['Editor'] = df_merged.get('Editor')
df_merged['Publisher'] = df_merged.get('Publisher')
df_merged['Year of Publication'] = df_merged.get('Year of Publication')
df_merged['Language'] = df_merged.get('Language')

# Drop duplicates
df_merged.drop(columns=['Author_x', 'Author_y'], inplace=True)

# Drop File Name if not needed
df_merged.drop(columns=['File Name'], inplace=True)

# Fine tune Title and Author
df_merged['Title'] = df_merged['Title'].apply(lambda x:x.split('_')[0])
df_merged['Author'] = df_merged['Author'].apply(lambda x:x.split('_')[0])
```
